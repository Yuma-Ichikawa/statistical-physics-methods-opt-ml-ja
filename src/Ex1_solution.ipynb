## Exercise 1.4 - Metropolis-Hastings algorithm for Curie-Weiss

初期配位は, $S(0)= (1,\dots, 1)$とする. 

また, Curie-Weiss模型のHamiltonian
$$\mathcal{H}_{N}(S) = - \frac{1}{2N} \sum_{ij} S_{i} S_{j} - h \sum_{i} S_{i}$$

の場合, $\Delta E = E_{\mathrm{now}} - E_{\mathrm{flip}}$は以下のように評価できる.

\begin{align} E_{\mathrm{now}} - E_{\mathrm{flip}} &= - N \left(\frac{1}{2} \bar{S}^{2} + h \bar{S}  \right) - \frac{N}{2} \left(\bar{S} - \frac{2S_{i}}{N}  \right)^{2} - N h \left(\bar{S} - \frac{2 S_{i}}{N}  \right) \\ &= 2S_{i}\left(\bar{S} - \frac{1}{N} S_{i} + h  \right) \end{align}

```python
def metropolis_update(S: np.ndarray, beta: float, h: float) -> np.ndarray:
    """一回のMetropolis Update
    
    Parameters
    ----------
    S : ndarray of shape (N, )
        spin configuration at time t
    beta : float
           inverse temperature
    h : float
        magnetic field 

    Returns
    ---------
    S : ndarray of shape (N, )
        spin configuration at time t+1
    """
    
    # randomにスピンを選択
    i = np.random.randint(low=0, high=len(S)) 
    # エネルギー差を計算
    dE = 2*S[i]*(np.mean(S) -S[i]/len(S) + h)  
    # accept or reject
    if np.random.random() < np.exp(-beta * dE):
        S[i] = -S[i] 
    return S
```

```python
def MH_MCMC(beta, h, S0, steps = 10000) -> np.ndarray:
    """Metropolis-Hasting MCMC
    
    Parameters
    ----------
    beta : float
           inverse temperature
    h : float
        magnetic field 
    S0 : ndarray of shape (N, )
        initial configuratioin 
    num_step : int
                number of steps

    Returns
    ---------
    S : ndarray of shape (N, )
        spin configuration at time t+1
    """
    S = S0.copy()
    m = np.zeros(0)
    for t in np.arange(steps-1):
        m = np.hstack([m, np.mean(S)])
        S = metropolis_update(S, beta, h)

    return m
```

```python
# 理論値を取得する関数
def H(m):
    return - 0.5*((1+m) * np.log((1+m)/2) + (1-m)*np.log((1-m)/2))

def potential(m, beta, h):
    return 0.5*beta*(m**2) + beta*h*m + H(m)

def get_mstar(beta, h):
    ms = np.linspace(-0.999, 0.999, 1000)
    pot = potential(ms, beta, h)
    # peakを持つindexを取得
    peaks = find_peaks(pot)[0]
    return [ms[peak] for peak in peaks]
```

```python
h = 0
beta = 1.2
mstars = get_mstar(beta, h)

print('Minima at m={}'.format(mstars))

ms = np.linspace(-0.999,0.999, 100)
pot = potential(ms, beta, h)

fig, ax = plt.subplots()
ax.plot(ms, pot, label=r'$\beta={}$'.format(beta))

for mstar in mstars:
    ax.scatter([mstar], [potential(mstar, beta, h)], marker='o', color='black')

ax.set_xlabel('m')
ax.set_ylabel(r'$\phi(m)$')

plt.show()
```

```python
h = 0
beta = 1.2

fig, ax = plt.subplots(5, 1, figsize=(10, 16))
for i, N, in enumerate([10, 50, 100, 200, 1000]):
    # initial spins
    S0 = np.ones(N)
    ms = MH_MCMC(beta, h, S0, steps = 100*N)
    
    ax[i].plot(ms[:-1])
    for mstar in mstars:
        ax[i].axhline(mstar, color='k', ls='--')

    ax[i].set_xlabel('steps')
    ax[i].set_ylabel(r'magnetisation')
```

$\beta = 1.2$, $h=0$では, $\phi(m)$に2つの対称な大域的最適解が存在する. この2つの値は, 系の平衡状態の値に対応する. Metropolis-Hastingsアルゴリズムの各ステップでは, 1つのスピンをさせることを思い出すと, 磁化$m^{\ast}$と磁化$-m^{\ast}$の間の典型的な時間は$t \sim e^{N}$のオーダーである. アルゴリズムを$\bar{S} = (1, \ldots, 1)$で初期化しているため, $m^{\ast}$に近く, $N \to \infty$では磁化$m^{\ast}$に到達する. しかし, 数値的には常に有限の$N$となるため, $t=e^{N}$だけアルゴルリズムを実行すると$-m^{\ast}$に到達する可能性もある. 実際, $N=10,50$の場合, そのような現象が見られる(有限サイズ効果). 

In this case, the free energy potential $\phi(m)$ has only one global maximum. Let's run exactly the above for this case.

```python
h = -0.1
beta = 1.2
mstars = get_mstar(beta, h)

print('Minima at m={}'.format(mstars))

ms = np.linspace(-0.999,0.999, 100)
pot = potential(ms, beta, h)

fig, ax = plt.subplots()
ax.plot(ms, pot, label=r'$\beta={}$'.format(beta))

for mstar in mstars:
    ax.scatter([mstar], [potential(mstar, beta, h)], marker='o', color='black')

ax.set_xlabel('m')
ax.set_ylabel(r'$\phi(m)$')

plt.show()
```

## Exercise 1.5 - Glauber algorithm for Curie-Weiss

Our second example of algorithm to sample from $P_{\beta}$ is the Glauber algorithm. The idea here is to follow an update rule mapping equilibrium configurations to equilibrium configurations. Again, the starting point is a fixed configuration $S^{0}$, which for concretness we will take to be $S^{0} = (1,\cdots, 1)$. The update rule is then given by:
1. Let $S^{t}$ be the configuration at time $t\geq 0$, and let $m^{t} = \frac{1}{N}\sum\limits_{i=1}^{N}S^{t}_{i}$ be the magnetisation associated to it.
2. Pick a random component $i\in\{1,\cdots, N\}$ from $S^{t}$.
3. Draw a random $p$ number uniformly in $[0,1]$. Update the configuration following the rule:
\begin{align}
S_{i}^{t+1} = 
\begin{cases}
+1 & \text{ if } p < \frac{1}{2}\left[1+\tanh(\beta(m+h))\right] \\
-1 & \text{otherwise} 
\end{cases},&& S^{t+1}_{j} = S^{t}_{j} \qquad \text{ for } j\neq i
\end{align}

4. Go back to step 1 with $t\mapsto t+1$

One of the great advantages of the Glauber algorithm is that we have an *analytical* differential equations describing the evolution the average magnetisation $\langle m^{t}\rangle$ as a function of $t$ for $N\gg 1$. This is derived in the first part of the exercise, and is given by:

\begin{align}
\partial_{t} \langle m \rangle = -\langle m \rangle+\tanh(\beta(\langle m \rangle+h)) 
\end{align}

Therefore, we can track the evolution of the algorithm and compare with the expected trajectory at $n\to\infty$. Numerically, we can simulate this ODE by using a Euler discretisation (one of the many possible solutions):

\begin{align}
m^{t+1} = m^{t} + \delta \left[-m^{t}+\tanh(\beta( m^{t}+h)) \right], && \delta\ll 1
\end{align}


As before, let's start implementing the update rule $S^{t}\mapsto S^{t+1}$ for the Glauber algorithm and for our ODE $m^{t}\to m^{t+1}$

```python
def glauber_update(S: np.ndarray, beta: float, h: float) -> np.ndarray:
    """一回のGlauber Update
    
    Parameters
    ----------
    S : ndarray of shape (N, )
        spin configuration at time t
    beta : float
           inverse temperature
    h : float
        magnetic field 

    Returns
    ---------
    S : ndarray of shape (N, )
        spin configuration at time t+1
    """

    m = np.mean(S) 
    i = np.random.randint(low=0, high=len(S)) 
    if np.random.random() < (1+np.tanh(beta*(m+h)))/2:
        S[i] = +1
    else:
        S[i] = -1
        
    return S

def ode_update(m: float, beta: float, h: float, step_size = 1e-3) -> float:
    """一回のMetropolis Update
    
    Parameters
    ----------
    S : ndarray of shape (N, )
        spin configuration at time t
    beta : float
           inverse temperature
    h : float
        magnetic field 
    step_size : float
                time step 

    Returns
    ---------
    magnetisation at time t+1 : float
    """
    return m + step_size*(-m + np.tanh(beta*(m+h)))
```

Now we can implement a function to iterate the updates

```python
def iterate_glauber(beta, h, S0, num_step=1000) -> np.ndarray:
    """Glauber Dynamics
    
    Parameters
    ----------
    beta : float
           inverse temperature
    h : float
        magnetic field 
    S0 : ndarray of shape (N, )
        initial configuratioin 
    num_step : int
                number of steps

    Returns
    ---------
    S : ndarray of shape (N, )
        spin configuration at time t+1
    """
    S = S0.copy()
    m = np.zeros(0)
    for t in np.arange(num_step-1):
        m = np.hstack([m, np.mean(S)])
        S = glauber_update(S, beta, h)
        
    return m

def iterate_glauber_ode(beta, h, m0=1, num_step=1000, step_size=1e-3) -> np.ndarray:
    """Glauber Dynamics ODE
    
    Parameters
    ----------
    beta : float
           inverse temperature
    h : float
        magnetic field 
    m0 : float
        initial magnetization 
    num_step : int
                number of steps
    step_size : float
                time step

    Returns
    ---------
    trajectory of the magnetisation m(t) : ndarray
    """
    mt = m0
    m_memory = np.zeros(0)
    for t in np.arange(num_step-1):
        m_memory = np.hstack([m_memory, mt])
        mt = ode_update(mt, beta, h, step_size=step_size)
        
    return m_memory
```

```python
# Case 1:
beta = 1.2
h = 0

mstars = get_mstar(beta, h)

fig, ax = plt.subplots(5, 1, figsize=(10,18))
ax[0].set_title(r'Glauber dynamics for $\beta=1.2$ and $h=0$')

for i, N, in enumerate([10, 50, 100, 200, 1000]):
    S0 = np.ones(N) # Start from all ones configuration
    m0 = np.mean(S0)
    
    ms = iterate_glauber(beta, h, S0, num_step = 100*N)
    ms_ode = iterate_glauber_ode(beta, h, m0, num_step = 100*N, step_size=1/N)
    
    ax[i].plot(ms[:-1], color='b', label='MCMC')
    ax[i].plot(ms_ode[:-1], color='r', label='ODE')

    for mstar in mstars:
        ax[i].axhline(mstar, color='k', ls='--', label='Theory')
        
    ax[i].set_xlabel('steps')
    ax[i].set_ylabel(r'magnetisation')
    ax[i].set_ylim([-1,1])
```
